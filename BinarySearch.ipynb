## Solving problem from linear search

```python

def locate_card(cards,query):
    location = 0
    while location<len(cards):
        if cards[location] == query :
            return location
        location+=1
        
    return -1
```

```python
def binarysearch(cards,query):
    low,high = 0,len(cards)-1
    
    while(low<=high):
        mid = (low+high)//2
        midvalue = cards[mid]
        
       # print("low:",low,"high:",high,"mid:",mid)
        
        if midvalue == query:
            while(True):
                if (cards[mid-1]== query):
                    mid=mid-1
                break
            return mid
        elif midvalue > query :
            low = mid + 1
        elif midvalue < query:
            high = mid-1
    
    return -1
```

## Home work problem

### determine how many times the  roated sorted list is rotated

**Test cases**

```python
test1 = {'input' :{ 'nums' :[4,1,2,3]},'output' : 1} #list of size 4 rotated 1 time
test2 = {'input' :{ 'nums' :[4,5,6,7,8,1,2,3]},'output' : 5} # list of size 8 rotated 5 times 
test3 = {'input' :{ 'nums' :[4,6,8,9,10]},'output' : 0}#list of size 5 not rotated/rotated n times
test4 = {'input' :{ 'nums' :[]},'output' : 0} # empty list
test5 = {'input':{'nums':[1]},'output' : 0} # list of size 1 
test_case = [test1,test2,test3,test4,test5]
print(test_case)
```

**Solving the problem using Linear search**

```python
evaluate_test_cases(count_rotations_linear,test_case)
```

*Complexity of count_rotations_linear is `0(n)`*

```python
def count_rotations_binary(nums):
    left,high = 0,len(nums)-1

    # returns 0 if the list is in sorted order
    if nums[left] < nums[high]:
        return 0

    while (left<=high):
        mid = (left+high)//2

        if(mid < high and nums[mid]>nums[mid+1]):
            return mid+1
        if(mid > left and nums[mid-1]>nums[mid]):
            return mid
        if(nums[left]>nums[mid]):
            high = mid-1
        else:
            left = mid+1
    return 0
```

```python
evaluate_test_cases(count_rotations_linear,test_case)
```
