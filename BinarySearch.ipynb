## Solving problem from linear search

```python

def locate_card(cards,query):
    location = 0
    while location<len(cards):
        if cards[location] == query :
            return location
        location+=1
        
    return -1
```

```python
def binarysearch(cards,query):
    low,high = 0,len(cards)-1
    
    while(low<=high):
        mid = (low+high)//2
        midvalue = cards[mid]
        
       # print("low:",low,"high:",high,"mid:",mid)
        
        if midvalue == query:
            while(True):
                if (cards[mid-1]== query):
                    mid=mid-1
                break
            return mid
        elif midvalue > query :
            low = mid + 1
        elif midvalue < query:
            high = mid-1
    
    return -1
```

```python
evaluate_test_cases(binarysearch,test_case)
```
