1920.Build Array from Permutation

```python
def BuildArray(nums):
    return [nums [i] for i in nums]

BuildArray([0,2,1,5,3,4])
```

2011 – Final Value of Variable After Operations

```python
def AfterOperations(operations):
    val=0
    for i in operations:
        if i== "--X" or i=="X--":
            val-=1
        else:
            val+=1
    return val
AfterOperations(["X++","++X","--X","X--"])
```

1480 – Running Sum

```python
def runningSum(nums):
        
        arr = [0]*len(nums)
        sum = 0
        for  i in range(len(nums)):
            sum+=nums[i]
            arr[i]= sum
        return arr
runningSum([1,2,3,4])
```

LC 1365 – Smaller Numbers Than Current

```python
def smallerNumbersThanCurrent(nums):
        sorted_array = sorted(nums)
        unique = {}
        for i,num in enumerate(sorted_array):
            if num not in unique:
                unique[num] = i
        return [unique[x] for x in nums]
smallerNumbersThanCurrent([8,1,2,2,3])
```

LC 1512 – Number of Good Pairs

```python
from collections import Counter
nums = [1,2,2,3,3,3,4]
def numIdenticalPairs(nums):
        freq = Counter(nums)
        ans = 0
        for k in freq.values():
            ans+=k*(k-1)//2
        return ans
numIdenticalPairs(nums)
```

1768. Merge Strings Alternately

```python
def mergeAlternately(word1, word2):
    i=j=0
    result = []
    while i<len(word1) or j <len(word2):
        if i <len(word1):
            result.append(word1[i])
            i+=1
        if j <len(word2):
            result.append(word2[j])
            j+=1
    return "".join(result)
mergeAlternately("abc","pqrstu")
```

LC 485 – Max Consecutive Ones

```python
def findMaxConsecutiveOnes(nums):
        current_count = 0
        max_count=0
        for num in nums:
            if num==0 :
                current_count=0
            else:
                current_count+=1
                if current_count > max_count:
                    max_count = current_count
        return max_count
findMaxConsecutiveOnes([1,1,0,1,1,1])
```

LC 1295 – Find Numbers with Even Number of Digits

```python
import math
def findNumbers(nums):
    ans = 0
    for n in nums:
        if(int(math.log10(n)) + 1) % 2 == 0:
            ans += 1
    return ans
findNumbers([122,3432,12,124])
```

1071. Greatest Common Divisor of Strings approach for this

```python
def gcdStrings(str1,str2):
    if(str1+str2!=str2+str1):
        return ""
    def gcd(m,n):
        while(n):
            m,n = n,m%n
        return m
    n = gcd(len(str1),len(str2))
    return str1[:n]
gcdStrings('abcabcabc','abcabc')
```

1431. Kids With the Greatest Number of Candies

```python
def kidsWithCandies(candies, extraCandies):
    max_candies = max(candies)
    result = []
    for candy in candies:
        if candy + extraCandies >= max_candies:
            result.append(True)
        else:
            result.append(False)
    return result
kidsWithCandies([2,3,5,1,3],3)
```
