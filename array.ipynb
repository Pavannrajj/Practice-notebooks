1.Given arr = [2, 7, 4, 1, 9]find:

a. The sum
b. The maximum value

```python
arr = [2,7,4,1,9]
arr
```

```python
# sum
sum = 0
for i in range(len(arr)):
    sum+=arr[i]
print("sum of array : {}".format(sum))
```

```python
#maximum

max_val = arr[0]
for i in range(1,len(arr)):
    if arr[i] > max_val:
        max_val = arr[i]
print("max_val : {}".format(max_val))
```
