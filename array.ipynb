```python
arr = [2,7,4,1,9]
arr
```

```python
# sum
sum = 0
for i in range(len(arr)):
    sum+=arr[i]
print("sum of array : {}".format(sum))
```

```python
#maximum

max_val = arr[0]
for i in range(1,len(arr)):
    if arr[i] > max_val:
        max_val = arr[i]
print("max_val : {}".format(max_val))
```

**Frequency counting**

```python
arr = [1,2,3,4,5,6,7,7,4,3,3,5]
freq = {}

for num in arr:
    if num not in freq:
        freq[num] = 1
    freq[num] +=1

print("frequency : ",freq)
```

**Use two pointers to reverse:**
`[1, 2, 3, 4, 5]`

```python
arr = [1,2,3,4,5]
low,high = 0,len(arr)-1
while(low<high):
    arr[low],arr[high] = arr[high],arr[low]
    low+=1
    high-=1
print("reversed arr1:",arr)
```

**Use slow-fast pointers:**
Move zeros in:`[0, 1, 0, 3, 12]`

```python
arr=[0,1,0,3,12]

slow = 0

for fast in arr:
    if fast!=0:
        arr[slow]=fast
        slow+=1
for fast in range(slow,len(arr)):
    arr[slow]=0
    slow+=1
print(arr)
```

Variable sliding window:
Find longest subarray with sum â‰¤ `7`
`Array = [2, 1, 5, 1, 3]`

Template

```python
# left = 0
# for right in range(n):
#     # 1. include arr[right] in window

#     while condition is violated:
#         # shrink from left
#         left += 1

#     # 2. update answer
```

```python
arr = [2,1,5,1,3]

left = 0
window_sum = 0
best = 0
for right in range(len(arr)):
    window_sum += arr[right]

    while(window_sum > 7):
        window_sum-=arr[left]
        left+=1
    best = max(best,right-left+1)

print("best:",best)
```
