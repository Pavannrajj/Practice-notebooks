### Pattern printing

**`1`. Right-angle triangle (stars)**

```python
def right_angle_triangle(n):
    for i in range(1,n+1):
        print('*'*i)
right_angle_triangle(5)
```

**`2`. Inverted right-angle triangle (stars)**

```python
def inverted_right_angle_triangle(n):
    for i in range(n,0,-1):
        print('*'*i)
inverted_right_angle_triangle(5)
```

**`3`. Centered pyramid (stars)**

```python
def centered_pyramid(n):
    for i in range(n+1):
        spaces=' '*(n-i)
        stars = '*'*(2*i-1)
        print(spaces+stars)
centered_pyramid(4)
```

**`4`. Inverted centered pyramid (stars)**

```python
def inverted_centered_pyramid(n):
    for i in range(n,0,-1):
        spaces = " "*(n-i)
        stars =  "*"*(2*i-1)
        print(spaces,stars)
inverted_centered_pyramid(4)
```

**`5`.Number triangle (increasing numbers per row)**

```python
def number_triangle(n):
    for i in range(1,n+1):
        print(''.join(str(j) for j in range (1,i+1)))
number_triangle(4)
```
